```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/SuperMarket Analysis.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(x='Payment', data=df)
plt.title('Payment Method Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Product line', y='gross income', data=df)
plt.xticks(rotation=45)
plt.title('Gross Income by Product Line')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(),
            annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Rating', y='Sales', data=df)
plt.title('Sales vs Rating')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

daily_sales = df.groupby('Date')['Sales'].sum()

plt.figure(figsize=(10,4))
daily_sales.plot()
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(x='Branch', data=df)
plt.title('Number of Sales per Branch')
plt.show()
```
